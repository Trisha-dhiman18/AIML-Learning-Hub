**Step 1: Import Pandas**

First we import the pandas library.

```python
import pandas as pd
```

Pandas is used for data analysis and table-like data (DataFrame).

**Step 2: Create the Data**



```python
data = {
    "product_id": [0, 1, 2, 3, 4],
    "low_fats": ["Y", "Y", "N", "Y", "N"],
    "recyclable": ["N", "Y", "Y", "Y", "N"]
}
```

**Step 3: Convert Data into DataFrame**

```python
products = pd.DataFrame(data)
```

**Step 4: Apply Filtering Condition**

```python
filtered_products = products[
    (products["low_fats"] == "Y") & 
    (products["recyclable"] == "Y")
]
```

**Step 5: Select Only Required Column**

```python
result = filtered_products[["product_id"]]
```

**Step 6: Display the Result**

```python
print(result)
```
