## Rearrange Products Table

**Step 1: Import Pandas**

```python
import pandas as pd
```

**Step 2: Create Data**

```python
data = {
    "product_id": [0, 1],
    "store1": [95, 70],
    "store2": [100, None],
    "store3": [105, 80]
}
```

**Step 3: Create DataFrame**

```python
products = pd.DataFrame(data)
```

**Step 4: Apply Melt()** <br>
**melt()**	Convert wide → long
id_vars	Column to keep fixed
value_vars	Columns to unpivot

```python
result = products.melt(
    id_vars=["product_id"],
    value_vars=["store1", "store2", "store3"],
    var_name="store",
    value_name="price"
)
```

**Step 5: Remove Null Values**

```python
result = result.dropna()
```

**Step 6: Print Result**

```python
print(result)
```

**LeetCode Style**

```python
import pandas as pd

def rearrange_products_table(products: pd.DataFrame) -> pd.DataFrame:
    
    result = products.melt(
        id_vars=['product_id'],
        value_vars=['store1', 'store2', 'store3'],
        var_name='store',
        value_name='price'
    )
    
    return result.dropna()
```
