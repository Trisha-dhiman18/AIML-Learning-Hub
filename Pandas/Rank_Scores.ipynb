## Rank Scores

**Step 1: Import Pandas**

```python
import pandas as pd
```

**Step 2: Create Data**

```python
data = {
    "id": [1,2,3,4,5,6],
    "score": [3.50,3.65,4.00,3.85,4.00,3.65]
}
```

**Step 3: Create DataFrame**

```python
scores = pd.DataFrame(data)
```

**Step 4: Apply Dense Ranking**

```python
scores["rank"] = scores["score"].rank(method="dense", ascending=False).astype(int)
```

**Step 5: Select Required Columns**

```python
result = scores[["score", "rank"]].sort_values(by="score", ascending=False)
```

**Step 6: Print Result**

```python
print(result)
```

**LeetCode Style**

```python
import pandas as pd

def order_scores(scores: pd.DataFrame) -> pd.DataFrame:
    
    scores['rank'] = scores['score'].rank(method='dense', ascending=False).astype(int)
    
    return scores[['score', 'rank']].sort_values(by='score', ascending=False)
```
