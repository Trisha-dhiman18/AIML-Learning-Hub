## Patients with Condition

**Step1 - import pandas**

```python
import pandas as pd
```

**Step2 - Create Data**

```python
data = {
    "patient_id": [1,2,3,4],
    "patient_name": ["Alice","Bob","Charlie","David"],
    "conditions": [
        "DIAB100 HTN",
        "HTN ASTHMA",
        "ASTHMA DIAB200",
        "ADIAB100"
    ]
    }
```

**Step3 - Create DataFrame**

```python
patients = pd.DataFrame(data)
```

**Step4 - Apply Condition**

```python
result = patients[
    patients["conditions"].str.contains(r'(^|\s)DIAB1')
]
```

**Step5 - Print Result**

```python
print(result)
```

**LeetCode Style**

```python
import pandas as pd

def find_patients(patients: pd.DataFrame) -> pd.DataFrame:
    
    return patients[
        patients['conditions'].str.contains(r'(^|\s)DIAB1')
    ]
```
