## Article Views

**Step 1: Import pandas**

```python
import pandas as pd
```

**Step 2: Create Data**

```python
data = {
    "article_id": [1,1,2,2,3],
    "author_id": [3,3,7,7,4],
    "viewer_id": [3,4,7,7,4],
    "view_date": ["2019-08-01","2019-08-02","2019-08-01","2019-08-02","2019-08-01"]
}
```

**Step 3: Create DataFrame**

```python
views = pd.DataFrame(data)
```

**Step 4: Apply condition**

```python
filtered = views[views["author_id"] == views["viewer_id"]]
```

**Step 5: Select column**

```python
result = filtered[["author_id"]]
```

**Step 6: Remove duplicates**

```python
result = result.drop_duplicates()
```

**Step 7: Rename column**

```python
result = result.rename(columns={"author_id": "id"})
```
