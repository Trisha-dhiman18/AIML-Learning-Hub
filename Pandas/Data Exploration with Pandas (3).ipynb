<div style="background-color: #DCECE9; color: black; padding: 4px;border-radius: 10px; text-align: center;">
    <h2>Visualising Data
</h3> </div>

```python
# Plotting Bar Chart for better visualisation.
import matplotlib.pyplot as plt

categories = data['Occupation']
values = data['Stress Level']
plt.bar(categories,values, edgecolor = 'grey')  # plt.bar() -> for bar chart and without it bar chart will not be plotted.


plt.title("Bar Chart")
plt.xlabel("Stress Level")
plt.ylabel("Occupation")
plt.show()
```

```python
# Plotting Pie Chart for better Visualisation
plt.figure(figsize = (8,8))
count = data["Occupation"].value_counts()
plt.pie(
    count.values,
    labels = count.index,
    autopct = '%0.0f%%',
    startangle = 90
)

plt.show()
    
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Quality of Sleep"], bins=15, color='skyblue', edgecolor='black')

# Customization
plt.title("Histogram Chart", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.grid(False)

plt.show()
```
