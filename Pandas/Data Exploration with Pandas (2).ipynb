<div style="background-color: #DCECE9; color: black; padding: 4px;border-radius: 10px; text-align: center;">
    <h2>Visualising Data
</h3> </div>

```python
# Plotting Bar Chart for better visualisation.
import matplotlib.pyplot as plt

categories = data['Food_Item']
values = data['Meal_Type']
plt.bar(categories,values, edgecolor = 'grey')  # plt.bar() -> for bar chart and without it bar chart will not be plotted.


plt.title("Bar Chart of Food Items by Meal type")
plt.xlabel("Food")
plt.ylabel("Meal Type")
plt.show()
```

```python
# Plotting Pie Chart for better Visualisation
plt.figure(figsize = (8,8))
count = data["Meal_Type"].value_counts()
plt.pie(
    count.values,
    labels = count.index,
    autopct = '%0.0f%%',
    startangle = 90
)

plt.show()
    
```

```python
# Plotting Histogram Chart for visualisation
plt.figure(figsize=(8, 5))
plt.hist(data["Meal_Type"], bins=15, color='skyblue', edgecolor='black')

# Customization
plt.title("Histogram of Meal Type", fontsize=14)
plt.xlabel("Females")
plt.ylabel("")
plt.grid(False)

plt.show()
```
