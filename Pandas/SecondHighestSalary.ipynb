**Step 1: Import pandas**

```python
import pandas as pd
```

**Step 2: Create data**

```python
data = {
    "id": [1,2,3,4],
    "salary": [100,200,300,200]
}
```

**Step 3: Create DataFrame**

```python
employee = pd.DataFrame(data)
```

**Step 4: Remove duplicate salaries & Sort descending**

```python
unique_salaries = employee["salary"].drop_duplicates().sort_values(ascending=False)
```

**Step 5: Check if second highest exists**

```python
if len(unique_salaries) >= 2:
    answer = unique_salaries.iloc[1]
else:
    answer = None
```

**Step 7: Create result DataFrame**

```python
result = pd.DataFrame({
    "SecondHighestSalary": [answer]
})
```

**Step 8: Print Result**

```python
print(result)
```
