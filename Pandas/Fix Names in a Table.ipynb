## Fix Names in a Table

**Step 1: Import pandas**

```python
import pandas as pd
```

**Step 2: Create data**

```python
data = {
    "user_id": [1,2],
    "name": ["aLice","bOB"]
}
```

**Step 3: Create DataFrame**

```python
users = pd.DataFrame(data)
```

**Step 4: Fix names**

```python
users["name"] = users["name"].str.capitalize()
```

**Step 5: Sort by user_id**

```python
result = users.sort_values(by="user_id")
```

**Step 6: Print result**

```python
print(result)
```
