<div style="background-color: #DCECE9; color: black; padding: 4px;border-radius: 10px; text-align: center;">
    <h2>What is Pandas ? 
</h3> </div>

Pandas is a **Python library** used for **data manipulation** and **analysis**.  
It works with **structured data** like tables (rows & columns).

```python
# Importing Pandas library
import pandas as pd

# Checking Pandas version
print("Pandas Version:", pd.__version__)
```

<div style="background-color: #DCECE9; color: black; padding: 4px;border-radius: 10px; text-align: center;">
    <h2> Pandas Series
</h3> </div>

A **Series** is a **1-dimensional labeled array**.  
It can hold data of any type (integer, string, float, etc.).

**Syntax:** <br>
pd.Series(data, index)


```python
# Creating a Series from a list
data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)
```

```python
# Creating a Series with custom index
s_custom = pd.Series(data, index=['A', 'B', 'C', 'D'])
print("Series with Custom Index:")
print(s_custom)
```

<div style="background-color: #DCECE9; color: black; padding: 4px;border-radius: 10px; text-align: center;">
    <h2>Pandas DataFrame
</h3> </div>

 A **DataFrame** is a **2-dimensional table** with rows and columns.  
It can be created from:
- Lists
- Dictionaries
- NumPy arrays
- CSV/Excel files

**Syntax:**<br>
pd.DataFrame(data)


```python
# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'City': ['Delhi', 'Mumbai', 'Kolkata']
}

df = pd.DataFrame(data)
df
```

<div style="background-color: #DCECE9; color: black; padding: 4px;border-radius: 10px; text-align: center;">
    <h2>Filtering Data
</h3> </div>

```python
import pandas as pd
df = pd.read_csv("airlines_flights_data.csv")
# Flights by specific airline
df[df['airline'] == "Air_India"]
```

```python
# Flights with arrival_time NOT equal to "Morning" and NOT equal to "Early_Morning"
df[(df['arrival_time'] != "Morning") & (df['arrival_time'] != "Early_Morning")]
```
