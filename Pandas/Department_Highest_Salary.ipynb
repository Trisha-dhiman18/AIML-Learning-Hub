```python
import pandas as pd

def department_highest_salary(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    
    merged = employee.merge(department, left_on='departmentId', right_on='id')
    
    max_salary = merged.groupby('departmentId')['salary'].transform('max')
    
    result = merged[merged['salary'] == max_salary]
    
    return result[['name_y', 'name_x', 'salary']].rename(
        columns={'name_y': 'Department', 'name_x': 'Employee', 'salary': 'Salary'}
    )
```

**Step 1: Import pandas**

```python
import pandas as pd
```

**Step 2: Create Employee table**

```python
employee_data = {
    "id": [1,2,3,4,5],
    "name": ["Joe","Jim","Henry","Sam","Max"],
    "salary": [70000,90000,80000,60000,90000],
    "departmentId": [1,1,2,2,1]
}
```

**Step 2: Create Employee table**

```python
department_data = {
    "id": [1,2],
    "name": ["IT","Sales"]
}

employee = pd.DataFrame(employee_data)
department = pd.DataFrame(department_data)
```

**Step 4: Merge both tables**

```python
merged = employee.merge(department, left_on="departmentId", right_on="id")
```

**Step 5: Find highest salary in each department**

```python
max_salary = merged.groupby("departmentId")["salary"].transform("max")
```

**Step 6: Filter employees having highest salary**

```python
filtered = merged[merged["salary"] == max_salary]
```

**Step 7: Select required columns**

```python
result = filtered[["name_y", "name_x", "salary"]].rename(columns={
    "name_y": "Department",
    "name_x": "Employee",
    "salary": "Salary"
})
```

**Step 8: Print Result**

```python
print(result)
```
