**1. Project Introduction**

Netflix is one of the world's largest streaming platforms with thousands of movies and TV shows.
In this project, we analyze the Netflix dataset to understand trends in content distribution.

**Objectives**

 We will answer the following questions:

- How many Movies vs TV Shows are available?

- Which countries produce the most content?

- What are the most popular genres?

- How has Netflix content grown over time?

- What is the most common rating category?

**2. Import Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Library	Purpose
Pandas	Data analysis
NumPy	Numerical operations
Matplotlib	Data visualization
Seaborn	Advanced visualizationm

**3. Load the Dataset**

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

**4. Explore Dataset Structure**

Check dataset shape.

```python
df.shape
```

**Meaning :**

- 8807 rows

- 12 columns

**5. Check Dataset Information**

```python
df.info()
```

**This shows:**

- column names

- data types

- missing values

**6. Check Missing Values**

```python
df.isnull().sum()
```

**Example missing columns:**
- director
- cast
- country
- date_added
- rating
- duration

**Fill missing values.**

```python
df['country'] = df['country'].fillna("Unknown")
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['date_added'] = df['date_added'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Unknown")
df['duration'] = df['duration'].fillna("Unknown")
```

```python
df.isnull().sum()
```

**7. Movies vs TV Shows**

Check distribution.

```python
df['type'].value_counts()
```

**Visualization:**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='type', data = df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

**Insight**

Netflix contains more movies than TV shows.

**8. Top Countries Producing Content**

```python
top_countries = df['country'].value_counts().head(10)
top_countries
```

**Insight**

USA produces the largest amount of Netflix content.

**9. Most Popular Genres**

```python
top_genres = df['listed_in'].value_counts().head(10)
top_genres
```

**Visualization:**

```python
plt.figure(figsize=(10,5))
top_genres.plot(kind='bar')
plt.title("Most Popular Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

**Insight**

Drama and International Movies are among the most common genres.

**10. Content Released Over Time**

```python
year_data = df['release_year'].value_counts().sort_index()
year_data
```

**Visualization:**

```python
plt.figure(figsize=(10,5))
year_data.plot(kind='line')
plt.title("Netflix Content Released Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

**Insight**

Netflix content increased rapidly after 2015.

**11. Rating Distribution**

Check ratings.

```python
df['rating'].value_counts()
```

**Visualization:**

```python
plt.figure(figsize=(10,5))
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title("Distribution of Content Ratings")
plt.show()
```

**Insight**

Most content is rated TV-MA and TV-14.

**12. Correlation Heatmap (Advanced Analysis)**

First select numeric columns.

```python
numeric_data = df.select_dtypes(include=['int64','float64'])
```

**Create Heatmap**

```python
plt.figure(figsize=(6,4))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**13. Key Insights**

 From our analysis we found:

1. Netflix has more movies than TV shows

2. USA produces the most content
 
3. Drama and international movies dominate

4. Netflix content increased significantly after 2015

5. Most content is rated TV-MA or TV-14
