## 16. Matrix Multiplication
Matrix multiplication is different from element-wise multiplication.
It is the core operation used in machine learning and deep learning models.

NumPy provides multiple ways to perform matrix multiplication.

```python
import numpy as np
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print(A * B)      # Element-wise multiplication
print(A @ B)      # Matrix multiplication
print(np.dot(A, B))
```

---

## 17. Dot Product
The dot product is used to multiply matrices and vectors.
It is widely used in linear regression and neural networks.

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print(np.dot(v1, v2))
```

---

## 18. Transpose of a Matrix
Transpose converts rows into columns and columns into rows.
It is commonly used to align matrix dimensions for multiplication.

```python
M = np.array([[1, 2, 3],
              [4, 5, 6]])

print(M.T)
```

---

## 19. Identity Matrix
An identity matrix is a square matrix with:
- 1s on the diagonal
- 0s elsewhere

It behaves like the number 1 in matrix multiplication.

```python
I = np.eye(3)
print(I)

print(A @ I[:2, :2])   # Identity effect

```

---

## 20. Inverse of a Matrix
The inverse of a matrix is used to solve linear equations.
Not all matrices have an inverse.

```python
C = np.array([[4, 7],
              [2, 6]])

C_inv = np.linalg.inv(C)
print(C_inv)

print(C @ C_inv)   # Should be close to identity
```

---

## 21. Stacking Arrays
Stacking is used to combine multiple arrays into one.
NumPy supports vertical and horizontal stacking.

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(np.hstack((x, y)))
print(np.vstack((x, y)))
```

---

## 22. Sorting and Searching
NumPy provides efficient functions for:
- Sorting data
- Finding indices
- Applying conditions

```python
data = np.array([40, 10, 30, 20])

print(np.sort(data))
print(np.argsort(data))
print(np.where(data > 25))
```
