## Zip
The **zip()** function in Python is used to combine multiple iterables such as lists or tuples into a single iterator of tuples. Each tuple contains elements taken from the iterables at the same index position. The iteration stops when the shortest iterable is exhausted. <br>
**Syntax**
zip(iterable1, iterable2, ...)

```python
a = [1, 2, 3]
b = ['a', 'b', 'c']

res = zip(a, b)
print(list(res))
```

```python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

print([list(row) for row in zip(*matrix)])

```

## Lambda
A **lambda** function is a small anonymous function used for simple operations without defining a full function using def. It can take any number of arguments but contains only a single expression.
<br>
**Syntax :**
lambda arguments: expression

```python
add = lambda x, y: x + y
print(add(5, 3))
```

```python
num = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(list(filter(lambda x: x % 2 == 0, num)))
```

## Filter
The **filter()** function is used to select elements from an iterable that satisfy a specific condition. It applies a function to each element and returns only those for which the function returns True.
<br>
**Syntax :**
filter(function, iterable)

```python
num = [1, 2, 3, 4, 5, 6]
res = filter(lambda x: x % 2 == 0, num)
print(list(res))
```

## Map
