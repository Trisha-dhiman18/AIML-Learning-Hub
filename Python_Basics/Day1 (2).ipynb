**1. Problem:** Write a Python function that accepts two integer numbers. If the product of the two numbers is less than or equal to 1000, return their product; otherwise, return their sum.

```python
def multiplication_or_sum(a, b):
    # Calculate product
    product = a * b
    
    # Check if product is within the threshold
    if product <= 1000:
        return product
    else:
        return a + b


result = multiplication_or_sum(20, 30)
print("The result is", result)


result = multiplication_or_sum(40, 30)
print("The result is", result)
```

**9. Problem:** Write a program to count the total number of vowels (a, e, i, o, u) present in a given sentence.

```python
s1 = input("Enter the string: ")
count = 0

for i in s1:
    if i.lower() in ('a', 'e', 'i', 'o', 'u'):
        count += 1

print("Total vowels =", count)
```

**12. Problem:** Write a function to return True if the first and last number of a given list is the same. If the numbers are different, return False.

```python

def check_nums(nums) :
    print("Given list: ",nums)
    first = nums[0]
    last = nums[-1]
    if first == last:
        return True
    else:
        return False
nums1 = [10, 20, 30, 40, 10]
print("Result: ",check_nums(nums1))
nums2 = [75, 65, 35, 75, 30]
print("Result: ",check_nums(nums2))
```

**16. Problem:** Write a program to check if a given number is a palindrome (reads the same forwards and backwards).

```python
def check_palindrome(num):
    print("Number: ",num)
    s1 = str(num)
    
    if s1 == s1[::-1]:
        print("The given number is palindrome.")
    else:
        print("The given number is not a palindrome.")
check_palindrome(121)
check_palindrome(142)
```

Problem: Start with a list of 10 numbers. Iterate through them and sort them into two separate lists: one for even numbers and one for odd numbers.

```python
numbers = [12, 7, 34, 21, 5, 10, 8, 3, 19, 2]
even_numbers = []
odd_numbers = []
for i in numbers:
    if i % 2 == 0:
        even_numbers.append(i)
    else:
        odd_numbers.append(i)
print(f"The list of even numbers is : {even_numbers} and the list of odd numbers is: {odd_numbers}")
```
