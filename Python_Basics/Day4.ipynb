**Problem: 61** Write a single-line list comprehension that takes a list of strings, filters out strings shorter than 4 characters, and converts the remaining strings to uppercase.

```python
words = ["apple", "bat", "cherry", "dog", "elderberry"]
new =[]
for i in words:
    if len(i) >= 4:
        new.append(i.upper())
print(f"Original : {words}")
print(f"Filtered Words :{new}")
        
```

**Problem: 63** Create a function that takes a string and returns a count of how many times each character appears. Ignore spaces and make it case-insensitive.

```python
from collections import Counter
def get_frequency(input_string):
    clear_text = input_string.lower().replace(" " , "")
    return Counter(clear_text)
text = "Python Programming"
freq = get_frequency(text)

print(f"Original: {text}")
print(f"Character Frequency: {freq}")
```

**Problem: 64** Write a function that determines if two strings are anagrams (contain the exact same characters in a different order).

**Problem: 65** Write a recursive function that takes a list containing other lists (of any depth) and returns a single “flat” list of all elements.

```python
def flatten(lst):
    flat_list =[]
    for item in lst:
        if isinstance(item,list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list
nested_data =[1, [2, 3], [4, [5, 6]], 7]
result = flatten(nested_data)

print(f"Original:  {nested_data}")
print(f"Flattened: {result}")
```
