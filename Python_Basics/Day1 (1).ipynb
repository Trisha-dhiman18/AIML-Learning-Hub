**1. Problem:** Write a Python function that accepts two integer numbers. If the product of the two numbers is less than or equal to 1000, return their product; otherwise, return their sum.

```python
def multiplication_or_sum(a, b):
    # Calculate product
    product = a * b
    
    # Check if product is within the threshold
    if product <= 1000:
        return product
    else:
        return a + b


result = multiplication_or_sum(20, 30)
print("The result is", result)


result = multiplication_or_sum(40, 30)
print("The result is", result)
```

**2. Problem:** Iterate through the first 10 numbers (0–9). In each iteration, print the current number, the previous number, and their sum.

```python
def state_tracking(a):
    numbers = range(a)
    prev = 0
    for i in range(0,a):
        sum = prev + i
        print(f"Current Number: {i} Previous Number: {prev} sum : {sum }")
        prev = i
state_tracking(10)
```

**6. Problem:** Write a program that calculates the factorial of a given number (e.g., 5!) using a for loop.

```python
n = 5
fact = 1
for i in range(1,n+1):
    fact = fact*i
print(fact)
```
