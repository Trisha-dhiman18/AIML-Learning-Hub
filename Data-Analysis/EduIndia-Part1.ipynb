**Exercise 19** – Using np.where() for Classification

```python
#. Use np.where() to create a column GENDER_GAP_LABEL that stores 'Male Higher' when male literacy exceeds female literacy, and 'Female Higher or Equal' otherwise.
df["GENDER_GAP_LABEL"] = np.where(
    df["MALE_LIT"] > df["FEMALE_LIT"],
    "Male Higher",
    "Female Higher or Equal"
)

df.head()
```

```python
# 75. Print value_counts() of this new column.
print(df["GENDER_GAP_LABEL"].value_counts())
```

```python
# 76. Compare this result to Exercise 19 - do the counts match?
print(df["GENDER_GAP_FLAG"].value_counts())

print(df["GENDER_GAP_LABEL"].value_counts())
```

**Exercise 20** – Basic Summary Report

```python
# 78. Using everything learned so far, print a short 6-line summary report using f-strings covering: total districts, 
# total states, national average literacy, state with highest literacy, state with lowest literacy, and total missing 
# values in the dataset.
# 79. Format all percentages to 2 decimal places.
# 80. This exercise has no charts - focus on clean, readable printed output.

total_districts = len(df)
total_states = df["STATNAME"].nunique()
average_literacy = df["OVERALL_LI"].mean()
state_average = df.groupby("STATNAME")["OVERALL_LI"].mean()
highest_state = state_average.idxmax()
lowest_state = state_average.idxmin()
missing_values = df.isnull().sum().sum()
print(f"Total Districts      : {total_districts}")
print(f"Total States         : {total_states}")
print(f"Average Literacy     : {average_literacy:.2f}%")
print(f"Highest Literacy     : {highest_state}")
print(f"Lowest Literacy      : {lowest_state}")
print(f"Missing Values       : {missing_values}")
```
