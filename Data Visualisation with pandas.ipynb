```python
# Plotting Bar Chart for better visualisation.
import matplotlib.pyplot as plt

categories = data['Below 18 years - Male']
values = data['Below 18 years - Female']
plt.bar(categories,values, edgecolor = 'grey')  # plt.bar() -> for bar chart and without it bar chart will not be plotted.


plt.title("Suicide Rate")
plt.xlabel("Male")
plt.ylabel("Female")
plt.show()
```

```python
# Plotting Pie Chart for better Visualisation
plt.figure(figsize = (8,8))
count = data["Below 18 years - Female"].value_counts()
plt.pie(
    count.values,
    labels = count.index,
    autopct = '%0.0f%%',
    startangle = 90
)

plt.show()
    
```

```python
plt.figure(figsize=(8, 5))
plt.hist(data["Below 18 years - Female"], bins=15, color='skyblue', edgecolor='black')

# Customization
plt.title("Histogram of Suicide detection", fontsize=14)
plt.xlabel("Females")
plt.ylabel("")
plt.grid(False)

plt.show()
```
