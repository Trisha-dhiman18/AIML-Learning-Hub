## 1ï¸âƒ£0ï¸âƒ£  Random Number Generation
Random numbers are used in machine learning for:
- Weight initialization
- Data shuffling
- Sampling
- Simulations

NumPy provides random number generation using `numpy.random`.

```python
import numpy as np
```

```python
# Random float number between 0 and 1
print(np.random.rand())
```

```python
# Random 5 floats between 0 and 1
print(np.random.rand(5))
```

```python
# Random floats for the 2D Matrix of size = 2,3
print(np.random.rand(2, 3))
```

```python
# Random integers
print(np.random.randint(1, 10)) # between 1 and 10
print(np.random.randint(1, 10, size=5))  # 1D Array
print(np.random.randint(1, 10, size=(2, 3)))  # 2D Array of size = 2,3
```

```python
# Setting random seed
np.random.seed(42)
print(np.random.rand(3))
```

## 1ï¸âƒ£1ï¸âƒ£ Boolean Masking & Filtering
Boolean masking is used to filter data based on conditions.
It creates a boolean (True/False) array and selects values where the condition is True.

This is heavily used in:
- Data cleaning
- Feature selection
- Conditional filtering in ML

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr > 25)
print(arr[arr > 25])
print(arr[(arr > 20) & (arr < 50)])
```

## 1ï¸âƒ£2ï¸âƒ£ Axis in NumPy
Axis defines the direction in which an operation is performed.

- `axis = 0` â†’ column-wise operation
- `axis = 1` â†’ row-wise operation

Axis is one of the most important and confusing concepts in NumPy.

```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

print(np.sum(matrix, axis=0))  # column-wise
```

```python
print(np.sum(matrix, axis=1))  # row-wise
```

```python
print(np.mean(matrix, axis=0))
```

```python
print(np.mean(matrix, axis=1))
```

```python
print(np.max(matrix, axis=0))
```

```python
print(np.min(matrix, axis=1))
```

## 1ï¸âƒ£3ï¸âƒ£ Broadcasting
- Broadcasting allows NumPy to perform operations on arrays of different shapes.

- The smaller array is automatically expanded to match the shape of the larger array.

- Broadcasting removes the need for loops and makes computation faster.

```python
arr1 = np.array([1, 2, 3])
print(arr1 + 10)

matrix2 = np.array([[1, 2],
                     [3, 4]])

print(matrix2 + np.array([10, 20]))
```

```python
# Scalar broadcasting
arr = np.array([1, 2, 3, 4])
print(arr * 5)
```

```python
# Row-wise broadcasting
mat = np.array([[1, 2, 3],
                [4, 5, 6]])

print(mat + np.array([10, 20, 30]))
```

```python
# Column-wise broadcasting
print(mat + np.array([[100], [200]]))
```

```python
# Reshape and Broad-Casting together
data = np.arange(1, 13)
data = data.reshape(3, 4)
print(data)

print(data + np.array([1, 2, 3, 4]))
```

## 1ï¸âƒ£4ï¸âƒ£ Basic Statistical Functions
NumPy provides built-in statistical functions such as:
- Mean
- Median
- Variance
- Standard Deviation

These functions are widely used in data analysis and machine learning.

```python
data = np.array([10, 20, 30, 40])

print(np.mean(data))
print(np.median(data))
print(np.var(data))
print(np.std(data))
```

```python
# Stats along axis
marks = np.array([[70, 80, 90],
                  [60, 75, 85],
                  [88, 92, 95]])

print(np.mean(marks, axis=0))
print(np.mean(marks, axis=1))
```

```python
# Feature matrix (X) and labels (y)
X = np.array([[1, 2],
              [2, 3],
              [3, 4],
              [4, 5]])

y = np.array([0, 0, 1, 1])
# Normalize features
X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
print(X_norm)

# Filter samples
print(X[y == 1])
```
